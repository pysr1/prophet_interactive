```python
from prophet_interactive import plotly_model
import pandas as pd 
import numpy as np 
from fbprophet import Prophet
import quandl
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
init_notebook_mode(connected=True)
```

```python
# The quandl package makes it easy to get data from FRED - just use the code obtained from FRED website
# comes right from URL https://fred.stlouisfed.org/series/GASREGCOVW

df = quandl.get("FRED/GASREGCOVW")
df = df.reset_index()
df.columns = ['ds', 'y']
m = Prophet()
m.fit(df)

# Python
future = m.make_future_dataframe(periods=52*2, freq = 'W')
# Python
forecast = m.predict(future)
iplot(plotly_model(df, forecast, point_estimate= False, trend = True))
```
